# EX

```c++
#include <iostream>
using namespace std;
int main(){
    int n;
    cin>>n;
    while(n>0){
        cout<<n<<' ';
        n--; //decrement
    }
    cout<<'\n'<<n<<'\n';
    
}
```

```c++
#include <iostream>
using namespace std;
int main(){
    for(int a=100, b=0; a>b; a--,b++)
    {
        cout<<a<<' 'b<<'\n';
    }
}
```

# For

```c++
for(int i=0; i<n;i++)
{
    cout<<i<<" ";
}

// i 技術變數
// i 起始直 0

for(變數出值);(迴圈持續條件),(每次回圈結束increment-decrement)
```
