## If 

```c++
#include<iostream>
using namespace std;
int main(){
    char cha;
    cout<<"請指定一種材料，我將用這種材料為你蓋出一座金字塔\n"
    "O:木材 |:竹子 H:鷹架\n"
    "你的選擇是:";
    here:
    cin>>cha;
    if(cha!='O'&&cha!='|'&&cha!='H')goto here;
    cout<<"  "<<cha<<'\n';
    cout<<" "<<cha<<cha<<cha<<endl;
    cout<<cha<<cha<<cha<<cha<<cha<<endl;
}
```

# 運算式

```c++
#include<iostream>
using namespace std;

int main()
{

int a=5, b=3, c=4;
cout<< a+b <<endl; //8
cout<< a-1 <<endl; //4
cout<< (a*a)/2 <<endl; //12.5 //not using float decimals will 88
cout<< ((a*a)-2)/2 <<endl; //11.5
a = b + c; // 7
cout<<a<<endl;
}
```
