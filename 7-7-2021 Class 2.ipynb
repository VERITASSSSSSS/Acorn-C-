## 輸入

```c++
使用者輸入、程式輸出
```

```c++
cout: 輸出 以<< 串接輸出資料
cin: 輸入  以>> 串接輸入資料
endl: 換行關鍵字，自動切換到下一行 (end of line)
```

```c++
int a=5, b=10; 
cout<<1<<endl;  //output int
cout<<1.5<<endl;//output float
cout<<'a'<<endl;//output char
cout<<"This is a string"<<endl; //output string
cout<<a<<" "<<b<<endl; //output variable a,b, set apart by blanks
```

```c++
#include <iostrean>
using namespace std;

int main(){
    int a=5, b=6;
    cout<<a<<''<<b<<endl;
    a=6,b=7;
    cout<<a<<""<<b<<endl;

}
```
